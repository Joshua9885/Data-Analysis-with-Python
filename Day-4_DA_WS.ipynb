```python
url1=("https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1")
```

```python
rq=requests.get(url1)
```

```python
forbes=BeautifulSoup(rq.text,"html")
forbes
```

```python
tab=forbes.find("table")
print(tab)
```

```python
thead=tab.find_all("th")
thead
```

```python
Headings=[i.text for i in thead]
Headings
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=Headings)
print(mydf)
```

```python
trows=tab.find_all('tr')
print(trows)
```

```python
tdata=tab.find_all('td')
print(tdata)
```

```python
data=[i.text for i in tdata]
print(data)
```

```python
print(data[0:4])
```

```python
ind=0
start=0
stop=4
for i in data:
  if stop <= len(data):
    mydf.loc[ind]=data[start:stop]
    start=start+4
    stop=stop+4
    ind=ind+1
mydf
```

```python
for i in range(len(mydf)):
  mydf.loc[i,"Name & India Rank"]=mydf.loc[i,"Name & India Rank"][3:]
mydf
```

```python
mydf.to_csv('forbes.csv')
```
