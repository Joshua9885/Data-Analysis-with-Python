```python
import seaborn as sns
```

```python
# titanic
ti=sns.load_dataset("titanic")
ti
```

```python
sns.relplot(data=ti,x='who',y='class',col='alive')
```

```python
sns.relplot(data=ti,x='who',y='class',hue='alive')
```

```python
sns.relplot(data=ti,x='who',y='class',hue='class',palette="bright")
```

```python
sns.relplot(data=ti,x='age',y='fare',kind="line")
```

```python
sns.catplot(data=ti,x="class",y="fare",kind="violin")
```

```python
sns.barplot(data=ti,x="class",y="alone",hue="sex")
```

```python
# iris
ir=sns.load_dataset("iris")
ir
```

```python
sns.relplot(data=ir,x="sepal_width",y="sepal_length",col="species")
```

```python
sns.relplot(data=ir,x="sepal_width",y="sepal_length",hue="species",palette="bright")
```

```python
sns.barplot(data=ir,x="petal_width",y="sepal_length",hue="species")
```

```python
sns.catplot(data=ir,x="petal_width",y="petal_length",hue="species",kind="violin")
```
