```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://en.m.wikipedia.org/wiki/List_of_tallest_buildings_in_India"
```

```python
req=requests.get(url)
```

```python
tall=BeautifulSoup(req.text,"html")
tall
```

```python
table=tall.find("table")
print(table)
```

```python
tabhead=table.find_all('th')
print(tabhead)
```

```python
heading=[i.text for i in tabhead]
print(heading)
```

```python
import pandas as pd
```

```python
mytall=pd.DataFrame(columns=heading)
print(mytall)
```

```python
tabrows=table.find_all('tr')
tabrows
```

```python
td=table.find_all('td')
td
```

```python
data=[i.text for i in td]
len(data)
```

```python
print(data[0:9])
```

```python
start=0
stop=9
for i in range(len(data)//9):
  mytall.loc[i]=data[start:stop]
  start=start+9
  stop=stop+9
mytall
```

```python
mytall.to_csv('Buildings.csv')
```
