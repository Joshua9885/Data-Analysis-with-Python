```python
import numpy as np
```

```python
from scipy import stats
```

```python
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
data1.mean()
data2.mean()
```

```python
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
print("-statistic:",t_stat)
print("P-value:",p_value)
```

```python
alpha=0.05
if p_value < alpha:
  print("Reject the null hypothesis. This means are significant")
else:
  print("Fail to reject the null hypothesis. This means are not signficant")
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Vaccine.csv')
```

```python
a=df['Efficiency'].mean()
print("Mean efficiency:",a)
b=df['Rec_Rate'].mean()
print("Mean recovery:",b)
```

```python
t_stat,p_value=stats.ttest_ind(a,b)
print("T-statistic:",t_stat)
print("P-value:",p_value)
```

```python
alpha=0.05
```

```python
if p_value > alpha:
  print("H0-realted")
else:
  print("H1-not realted")
```

```python
dfa=df[df['Vaccine']=='A']['Efficiency']
dfb=df[df['Vaccine']=='B']['Efficiency']
dfa
```

```python
#Annova testing
f_stat,p_value=stats.f_oneway(dfa,dfb)
print("F-statistic:",f_stat)
print("p-value:",p_value)
```

```python
alpha=0.05
if p_value > alpha:
  print("H0-realted")
else:
  print("H1-not realted")
```
