```python
#there are five subjects physcology,nano technology,medicine oncology and anthropology
#two students have applied to masters in any of these,with a line plot,show in which field they can excel. The out of for each subject is 70 scores
#studdent-1[67,45,73,71,57]
#student-2[43,67,685,3,85]
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
sub=["physcology","nanaotechnolgy","medicine","oncology","anthroplogy"]
plt.plot(sub,s1,label="student-1",color="c",linestyle=":",linewidth=4)
plt.plot(sub,s2,label="student-2",color="m")
plt.plot(sub,cut,label="cutoff",color="r")
plt.xlabel("sujects")
plt.ylabel("marks")
plt.legend()
```

```python
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
sub=["physcology","nanaotechnolgy","medicine","oncology","anthroplogy"]
plt.scatter(sub,s1,label="student-1",color="k",marker="h")
plt.scatter(sub,s2,label="student-2",color="grey")
plt.plot(sub,cut,label="cutoff",color="r")
plt.xlabel("sujects")
plt.ylabel("marks")
plt.legend()
```

```python
ind=np.array([25,51,84,131,160,189,220,250,267,298])
aus=np.array([15,41,94,110,151])
overs=np.arange(5,51,5)
plt.plot(overs,ind,label="India",color="b",marker=">",linewidth=3)
plt.plot(overs[:5],aus,label="Australia",color="y",marker="+")
plt.legend()
plt.title("Ind vs Aus")
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.grid()
```
